# Problem 1

```python
S = "BANANA"
S_length = len(S)
#print(S_length)
player1, player2 = 0,0

for i in range(S_length):
    if S[i] in "AEIOU":
        player1 += S_length - i
        #print("kevin",player1)
        
    else:
        player2 += S_length - i
        #print("stuart",player2)

        
if player1 > player2:
    print ("Kevin", player1)
elif player1 < player2:
    print ("Stuart", player2)
else:
    print ("Draw")
```
